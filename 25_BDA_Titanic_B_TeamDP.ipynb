# 타이타닉 : 사회·경제적 지위가 생존에 미친 영향은 어떤가?

| 변수명        | 의미               | 주석               |
| ---------- | ---------------- | ---------------- |
| `Pclass`   | 객실 등급 (1\~3)     | 사회적 계층 구분 기준     |
| `Fare`     | 운임 요금            | 경제 수준을 간접 반영     |
| `Embarked` | 탑승한 항구 (S, C, Q) | 지역적 특성/인프라 반영    |
| `Cabin`    | 객실 번호            | 유무 여부로 상류층 추정 가능 |

```python
# ttn = sns.load_dataset('titanic')
ttn = pd.read_csv('titanic1309.csv')
ttn.head()
```

```python
# 사망/생존 비율

labels = ttn['Survived'].value_counts().index
sizes = ttn['Survived'].value_counts().values

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Overall Survival Rate')
plt.legend(labels=['사망','생존'])
plt.show()
```

# 1️⃣ pclass (선실 등급)

```python
# 객실 등급 별 생존/사망 수

sns.countplot(x='Pclass', hue='Survived', data=ttn, palette='coolwarm')
plt.legend(labels=['사망', '생존'])
plt.title("객실 등급 별 생존/사망 수");
```

```python
plt.figure(figsize=(12, 5))



# 여성 - 등급 별 사망/생존 수
plt.subplot(1, 2, 1)
sns.countplot(data=ttn[ttn['Sex'] == 'female'], x = 'Pclass', hue='Survived', palette='Accent')
plt.yticks(np.arange(0, 500, 500))
plt.legend(labels=['사망','생존'])
plt.xlabel("female")
plt.ylabel("count")
plt.title("여성 - 등급 별 사망/생존 수")


# 남성 - 등급 별 사망/생존 수
plt.subplot(1, 2, 2)
sns.countplot(data=ttn[ttn['Sex'] == 'male'], x = 'Pclass', hue='Survived', palette='crest')
plt.yticks(np.arange(0, 500, 500))
plt.legend(labels=['사망', '생존'])
plt.xlabel("male")
plt.ylabel("count")
plt.title("남성 - 등급 별 사망/생존 수")


plt.show()
```

# 2️⃣ Embarked (탑승지)

| 값   | 의미          | 도시 이름 | 지역   |
| --- | ----------- | ----- | ---- |
| `S` | Southampton | 사우샘프턴 | 영국   |
| `C` | Cherbourg   | 셰르부르  | 프랑스  |
| `Q` | Queenstown  | 퀸스타운  | 아일랜드 |

```python
# 탑승지별 사망/생존 수

sns.catplot(x='Embarked',hue ='Survived', kind='count', data=ttn, palette='ocean')
plt.legend(labels=['사망', '생존'])
plt.title("탑승지별 사망/생존 수");
```

```python
# 탑승지별 객실 등급 분포 (막대그래프)

sns.countplot(data=ttn, x='Embarked', hue='Pclass', palette='summer')
plt.title("탑승지별 객실 등급 분포")
plt.xlabel("Embarked (Port)")
plt.ylabel("Passenger Count")
plt.legend(title="Pclass")
plt.show()
```

```python
# 탑승지별 객실 등급 비율 (히트맵)

heat_data = pd.crosstab(ttn['Embarked'], ttn['Pclass'], normalize='index')  # index 기준 정규화
plt.figure(figsize=(8, 5))
sns.heatmap(heat_data, annot=True, fmt=".2f", cmap='summer')
plt.title("탑승지별 객실 등급 비율")
plt.xlabel("Pclass")
plt.ylabel("Embarked")
plt.show()

# = C 항구의 1등석 비율이 0.51로 가장 많았고, Q 항구의 3등석 비율이 0.94로 가장 많음
```

```python
# 탑승지 + 객실등급별 생존률

grouped = ttn.groupby(['Embarked', 'Pclass', 'Survived']).size().reset_index(name='count')

# 각 (Embarked, Pclass) 조합 기준으로 정규화
grouped['percent'] = grouped.groupby(['Embarked', 'Pclass'])['count'].transform(lambda x: x / x.sum())

# x축 라벨 조합으로 묶기
grouped['label'] = grouped['Embarked'].astype(str) + ' - Class ' + grouped['Pclass'].astype(str)

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped, x='label', y='percent', hue='Survived', palette='Set2')

plt.title("탑승지 + 객실등급별 생존률")
plt.ylabel("Proportion")
plt.xlabel("Embarked - Pclass")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.legend(title="survived")

plt.tight_layout()
plt.show()
```

# 3️⃣ Deck (객실 층)

| Deck | 설명                        | 계층 추정        |
|------|-----------------------------|------------------|
| A    | 상층 갑판, 고급 객실        | 최상위층 (1등석) |
| B    | 고급 객실, 넓은 공간        | 1등석 중심       |
| C    | 중앙 상층, 좋은 위치        | 상류층           | 높음               |
| D    | 비교적 중간층, 접근성 좋음  | 1~2등석 혼합     |
| E    | 아래층, 중저가 객실         | 2등석 중심       |
| F    | 하층, 소음 많은 구역        | 3등석 중심       |
| G    | 가장 아래층, 제한적 사용    | 저소득층         |
| T    | 매우 특이한 객실(1개 존재)  | 불명             |
| NaN  | 객실 정보 없음              | 주로 3등석       |

```python
ttn['deck'] = ttn['Cabin'].str[0].replace('T', np.nan)
ttn['deck_initial'] = ttn['deck'].astype(str).str[0]
```

```python
ttn.groupby('deck_initial')['Survived'].mean().sort_values(ascending=False)
```

```python
# 객실 층별 사망/생존 수

sns.countplot(data=ttn, x='deck_initial', hue='Survived', palette='ocean')
plt.title("객실 층별 사망/생존 수")
plt.xlabel("deck_initial")
plt.ylabel("Passenger Count")
plt.legend(title="survived", labels=['사망', '생존'])
plt.show()

# = Nan(주로 3등석)이 사망자 수가 월등히 높음
```

```python
# 결측값 제거한 객실 층별 사망/생존 수

filtered = ttn[ttn['deck'].notna()]

# 그래프
sns.countplot(data=filtered, x='deck', hue='Survived', palette='ocean', order = ['A','B','C','D','E','F','G'])
plt.title("객실 층별 사망/생존 수(결측값 제외)")
plt.xlabel("Deck")
plt.ylabel("Passenger Count")
plt.legend(title="survived", labels=['사망', '생존'])
plt.show()
```

```python
# 객실 층별 생존/사망 수 (결측값 제외, 성별 분할)

filtered = ttn[ttn['deck'].notna()]

sns.catplot(data=filtered, x='deck', hue='Survived', col='Sex',
            kind='count', palette='cubehelix', height=5, aspect=1)

plt.subplots_adjust(top=0.8)
plt.suptitle("객실 층별 생존/사망 수 (결측값 제외, 성별 분할)")
plt.legend(title='survived',labels=['사망','생존'])
plt.show()
```

```python
# 객실 층별 성별 인원 수

sns.countplot(data=filtered, x='deck', hue='Sex', palette='cubehelix')
plt.title("객실 층별 성별 인원 수")
plt.xlabel("Deck")
plt.ylabel("Passenger Count")
plt.legend(title="성별", labels=["여성", "남성"])
plt.show()

# = 위와 아래 그래프를 비교해 보면, B층과 C층에 여성 승객의 비율이 높아 해당 층의 높은 생존률과 관련이 있는 것으로 보임.
```

# 4️⃣ Fare (운임 요금)

```python
# 운임 요금에 따른 생존률

ttn['Fare_bin'] = pd.qcut(ttn['Fare'], q=4, labels=['Low', 'Mid', 'High', 'Very High'])
sns.barplot(x='Fare_bin', y='Survived', data=ttn, hue = 'Fare_bin', palette= 'cool')
plt.title("운임 요금에 따른 생존률");
```

```python
# 선실 등급과 운임 요금의 관계

sns.boxplot(x='Pclass', y='Fare', data=ttn, hue='Pclass', palette='cool')
plt.title("선실 등급과 운임 요금의 관계");

# = 1등석 승객의 운임은 평균적으로 가장 높았고, 이상치도 많이 존재해 상류층 승객 사이에서도 가격 차이가 컸음을 알 수 있다. 반면 3등석은 전반적으로 운임이 낮고, 분포도 상대적으로 좁았음.
```

```python
ttn['survivedKor'] = ttn['Survived'].map({0:'사망', 1:'생존'})
```

```python
# 생존 여부에 따른 운임 분포


sns.violinplot(data=ttn, x='survivedKor', y='Fare', hue ='survivedKor', palette='cool', legend=True)
plt.legend(title = '생존여부')
plt.title("생존 여부에 따른 운임 분포")
plt.show()
# - 생존자와 사망자 각각의 운임 분포가 어떻게 다른지 확인할 수 있음
# - 운임이 높을수록 상류층일 가능성이 높고, 생존률도 높게 나타날 수 있음
```

```python
# 성별(female/male)에 따라 생존(Survived), 운임(Fare), 객실등급(Pclass) 변수들 간의 상관관계

female = ttn[ttn['Sex'] == 'female']
male = ttn[ttn['Sex'] == 'male']

# 히트맵 그리기 (1행 2열 subplot)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 여성
sns.heatmap(female[['Fare', 'Pclass', 'Survived']].corr(), annot=True,
            cmap='coolwarm', ax=axes[0])
axes[0].set_title("여성 상관계수")

# 남성
sns.heatmap(male[['Fare', 'Pclass', 'Survived']].corr(), annot=True,
            cmap='coolwarm', ax=axes[1])
axes[1].set_title("남성 상관계수")

plt.suptitle("성별에 따른 변수 간 상관관계 비교", fontsize=14)
plt.tight_layout()
plt.show()

# 파란색(cool)은 음의 상관관계, 붉은색(warm)은 양의 상관관계를 의미하며, 숫자가 1에 가까울수록 강한 상관관계를 나타냄
# - 여성의 경우 Fare(운임)과 Survived(생존) 간 양의 상관관계가 더 강하게 나타남.
# - 마찬가지로 남성도 Fare(운임)와 Survived 간의 양의 상관관계가 상대적으로 강하게 나타남을 알 수 있음.

#   → 결과적으로 운임이 높을수록 생존률 또한 높아지는 경향을 보여, Fare 변수는 생존 여부와 유의미한 관련이 있는 것으로 나타남.
```

## 1️⃣ 성별에 따른 생존자/사망자 수

```python
# 성별 분포 확인(탑승객 수) - 남성과 여성 비율 확인

# 그래프 크기 설정
plt.figure(figsize=(6, 4))

# 범주형 데이터의 개수를 세어 막대그래프로 표현
sns.countplot(x='Sex', data=tData, hue = 'Sex', palette=['#B5B2FF', '#FFB2D9'], legend = True)
plt.title('성별 분포 (전체 탑승객)')
plt.xlabel('성별')
plt.ylabel('탑승객 수')
plt.show()
```

```python
# 성별에 따른 생존/사망 분포 - 성별에 따른 생존자/사망자 수 달라짐 확인

plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=tData, palette=['#F15F5F', '#6799FF'])
plt.title('성별에 따른 생존/사망 분포')
plt.xlabel('성별')
plt.ylabel('인원 수')

# Survived 값이 0이면 사망, 1이면 생존이므로 명확하게 표기
plt.legend(title='생존 여부', labels=['사망(0)', '생존(1)'])
plt.show()
```

```python
# 성별 생존율 계산 - 성별에 따른 생존 확률이 어떻게 다른지 확
sex_surv_rate = tData.groupby('Sex')['Survived'].mean()
print("성별 생존율:\n", sex_surv_rate)

# 시각화
plt.figure(figsize=(6, 4))
sns.barplot(x=sex_surv_rate.index.map({'male': '남성', 'female': '여성'}), y=sex_surv_rate.values, palette=['#FFB2D9', '#B5B2FF'], hue = sex_surv_rate.index.map({'male': '남성', 'female': '여성'}))
plt.title('성별에 따른 생존율')
plt.xlabel('성별')
plt.ylabel('생존율')
plt.ylim(0, 1)
plt.show()
```

## 2️⃣ 나이 분포에 따른 생존자 비교

```python
# 전체 나이 분포 확인

plt.figure(figsize=(8, 4))

# 연속형 데이터의 분포를 히스토그램 형태로 표현
sns.histplot(tData['Age'].dropna(), bins=30, kde=False, color='#CEF279')
plt.title('전체 나이 분포')
plt.xlabel('나이')
plt.ylabel('탑승객 수')
plt.show()
```

```python
# 나이 구간별 생존율(연령대 분석) - 나이에 따른 생존자/사망자 분포 달라짐 확인

plt.figure(figsize=(8, 4))

# 생존자와 사망자 수를 쌓아서 누적 막대로 표현
sns.histplot(data=tData, x='Age', hue='Survived', multiple='stack', bins=30, palette={0: '#6799FF', 1: '#F15F5F'})
plt.title('나이 분포에 따른 생존자/사망자 비교')
plt.xlabel('나이')
plt.ylabel('탑승객 수')
plt.legend(title='생존 여부', labels=['사망(0)', '생존(1)'])
plt.show()
```

```python
# 나이 구간별 생존율 - 10세 단위로 나누고 연령대별 평균 생존율 확인

# Age를 나이 구간으로 분할하여 AgeGroup 열 생성
tData['AgeGroup'] = pd.cut(tData['Age'], bins=[0, 10, 20, 30, 40, 50, 60, 80],
                        labels=['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61+'])

# 그룹별 평균 생존율 계산 (Survived 열의 평균 = 생존율)
age_surv = tData.groupby('AgeGroup', observed=False)['Survived'].mean()

plt.figure(figsize=(8, 4))
sns.barplot(x=age_surv.index, y=age_surv.values,hue = age_surv.index, palette='rainbow')
plt.title('연령대에 따른 생존율')
plt.xlabel('연령대')
plt.ylabel('생존율')

# y축은 생존율이므로 0~1 사이로 고정
plt.ylim(0, 1)
plt.show()
```

*   0~10세 어린이 생존율이 가장 높음 → 구조 우선 정책의 영향을 강하게 받은 것으로 해석 가능
*   11~30세는 생존율이 급격히 낮아짐
*   노년층(61세 이상)은 생존율이 가장 낮음





## 3️⃣ 가족 정보 분석

```python
# 가족 수 변수 생성

tData['FamilySize'] = tData['SibSp'] + tData['Parch']
```

```python
# 가족 수 분포 - 가족 수 규모 확인

plt.figure(figsize=(8, 4))

# x축: FamilySize (형제자매(SibSp) + 부모자식(Parch)의 합)
warm_colors = ['#FFF5E1','#FFE4B5','#FFD39B','#FFC07C','#FFA94D','#FF9224','#FF7F50','#FF6347','#FF4500']
sns.countplot(x='FamilySize', data=tData, hue = 'FamilySize', palette=warm_colors)

plt.title('가족 수 분포')
plt.xlabel('가족 수 (형제자매/부모자식 합)')
plt.ylabel('탑승객 수')
plt.show()
```

```python
#가족 수에 따른 생존율 - 가족 수에 따른 탑승이 생존에 어떤 영향을 주는지 확인

# 가족 수별 평균 생존율 계산
# 0 = 혼자 탑승, 숫자가 클수록 가족 동반자 많음
family_surv = tData.groupby('FamilySize')['Survived'].mean()

plt.figure(figsize=(8, 4))
sns.lineplot(x=family_surv.index, y=family_surv.values, marker='o', color='#1E90FF')
plt.title('가족 수에 따른 생존율')
plt.xlabel('가족 수')
plt.ylabel('생존율')

# y축은 생존율이므로 0~1 범위
plt.ylim(0, 1)
plt.xticks(family_surv.index)
plt.show()
```

*   혼자 탑승한 승객은 생존 확률이 낮음
*   가족 수가 너무 적거나 너무 많으면 생존율 낮음 → 적당한 인원(1~3명)에서 가장 높은 생존율



## 4️⃣ 성별과 나이를 함께 분석

```python
# 성별과 나이에 따른 생존 분포 - 성별에 따라 어떤 나이대에서 생존/사망 발생했는지 확인

plt.figure(figsize=(8, 5))

# split=True: 생존/사망을 하나의 바이올린으로 반씩 나누어 표현
sns.violinplot(x='Sex', y='Age', hue='Survived', data=tData, split=True, palette={0: '#FF6B6B', 1: '#1E90FF'})
plt.title('성별과 나이에 따른 생존 분포')
plt.xlabel('성별')
plt.ylabel('나이')

# 범례 설정
plt.legend(title='생존 여부', labels=['사망(0)', '생존(1)'])
plt.show()
```

여성
*   전체적으로 생존의 밀도와 범위가 크다
*   다양한 연령대에서 생존자가 많이 나타남

남성
*   생존자 밀도가 적고 사망자 비중이 높음
*   생존 남성은 0~10세, 20대 초반 등 일부 연령대에서만 제한적으로 보임





## 5️⃣ 수치형 변수 간 상관관계 히트맵

```python
# 상관관계 분석을 위한 수치형 변수 목록 지

numeric_cols = ['Survived', 'Age', 'SibSp', 'Parch', 'Fare', 'FamilySize']

plt.figure(figsize=(8, 6))

# 히트맵 그리기
sns.heatmap(tData[numeric_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('수치형 변수 간 상관관계')
plt.show()
```

*   가족 수 관련 변수들은 구조적으로 연관되어 있어 매우 높은 양의 상관을 보임
*   운임과 생존 사이에는 약한 양의 상관관계 존재 -> 높은 운임일수록 생존 확률이 높음
*   가족이 있는 경우에 생존율이 높아지는 경향이 있음
*   나이와 생존율은 뚜렷한 상관관계는 없지만, 어린이 구조 우선순위 덕분에 높은 생존율을 가질 수 있음
*   전체적으로 상관계수 0.3 이하가 많아, 변수 간 선형 상관이 강하지 않음. 즉, 생존은 여러 복합 요인의 영향 결과임
